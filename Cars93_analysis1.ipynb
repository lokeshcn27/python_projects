```python
import pandas as pd
```

```python
df=pd.read_csv('Downloads/Cars93.csv')
```

```python
df.isna().sum().plot(kind='pie',autopct='%1.1f%%') # missing value's by visaulising
```

```python
# airbags contains 23.5% of null values
```

```python
for i in df.columns:
    if df[i].dtypes!=object:
        print(i)
        print()
        print(df[i].describe())
        print('******************************************************')
```

```python
# id,mpg.city,mpg.highway,enginesize,rpm,rev.per.mile have normal distrubution
```

```python
import seaborn as sns
```

```python
from matplotlib.pyplot import show
for i in df.columns:
    if df[i].dtypes!=object:
        sns.histplot(data=df,x=i,kde=True)
        print(i)
        show()
```

```python
# column names are normally distributed
```

```python
cat=[]
con=[]
for i in df.columns:
    if df[i].dtypes==object:
        cat.append(i)
    else:
        con.append(i)
```

```python
Xcat=df[cat]
Xcon=df[con]
```

```python
Xcon.corr() # checking correlations of all numeric columns with each other
```

```python
sns.heatmap(Xcon.corr(),annot=True)
```

```python
Xcon.corr()['Price'].sort_values(ascending=False).plot(kind='bar')
```

```python
sns.pairplot(data=Xcon)
```
